# 1、加载txt


```python
# 1、导入相关依赖
from langchain.document_loaders import TextLoader

# 2、定义TextLoader对象, file_path = '.txt的位置'
text_loader = TextLoader(file_path='./asset/load/01-langchain-utf-8.txt', encoding='utf-8')

# 3、加载
docs = text_loader.load()  # 返回List列表(Document对象)

# 4、打印
print(docs)

# 查看列表中元素的个数
print(len(docs))

# 查看列表中具体的元素内容
print(docs[0])

# 显示文档中的内容信息
print(docs[0].page_content)

# 显示Document对象的元数据
print(docs[0].metadata)
```

# 2、加载PDF


举例1：加载本地的文件

```python
# 1、导入相关的依赖包
from langchain.document_loaders import PyPDFLoader

# 2、定义PyPDFLoader
pdfLoader = PyPDFLoader(
    file_path='./asset/load/02-load.pdf'
)

# 3、加载
docs = pdfLoader.load()

# 4、展示结果
print(docs)

# 显示文本信息
print(docs[0])

# 显示元数据信息
print(docs[0].metadata)

# 显示具体的文本内容
print(docs[0].page_content)
```

举例2：加载网络中的文件

```python
# 1、导入相关的依赖包
from langchain.document_loaders import PyPDFLoader

# 2、定义PyPDFLoader
pdfLoader = PyPDFLoader(
    file_path="https://arxiv.org/pdf/2302.03803"
)

# 3、加载
docs = pdfLoader.load()

# # 4、展示结果
# print(docs)

# 显示长度
print(len(docs))

# 4、打印结果
for doc in docs:
    print(doc)
```

举例3：调用 load_and_split() 方法

```python
# 1、导入相关的依赖包
from langchain.document_loaders import PyPDFLoader

# 2、定义PyPDFLoader
pdfLoader = PyPDFLoader(
    file_path="./asset/load/02-load.pdf"
)

# 3、加载
docs = pdfLoader.load_and_split()  # 底层默认使用了递归字符文本且分器

# 4、展示结果
print(docs)

# 显示长度
print(len(docs))

# 4、打印结果
for doc in docs:
    print(doc)
```

# 3、加载CSV

```python
# 1、加载依赖包
from langchain_community.document_loaders.csv_loader import CSVLoader

# 2、定义CSVLoader
loader = CSVLoader(
    file_path='./asset/load/03-load.csv'
)

# 3、加载数据
data = loader.load()

# 4、数据展示
print(data)

# 其它项
print(type(data))
print(type(data[0]))
print(len(data))
print(data[0].page_content)
```

举例2：使用source_column参数指定文件加载的列, 保存到source变量中

```python
# 1、导入相关的依赖包
from langchain_community.document_loaders.csv_loader import CSVLoader

# 2、定义CSVLoader
loader = CSVLoader(
    file_path='./asset/load/03-load.csv',
    source_column='author'
)

# 3、加载数据
data = loader.load()

# 4、显示数据
print(data)

print(data[0])
print()
print(data[0].page_content)
print()
print(data[0].metadata)
```

# 4、加载JSON

举例1：加载json文件中的全部数据

```python
# 1、导入相关依赖包
from langchain_community.document_loaders import JSONLoader
from pprint import pprint

# 2、定义JSONLoader
json_loader = JSONLoader(
    file_path='./asset/load/04-load.json',
    jq_schema='.',  # 直接提取完整的JSON对象（包括所有字段）
    text_content=False  # 保持原始JSON结构, 将提取的数据转换为JSON字符串存入page_content字段中
)

# 3、加载数据
docs = json_loader.load()

# 4、显示数据
print(docs)
```

举例2：加载json文件中message[]中所有的content字段

```python
# 1、导入相关依赖包
from langchain_community.document_loaders import JSONLoader
from pprint import pprint

# 2、定义JSONLoader
json_loader = JSONLoader(
    file_path='./asset/load/04-load.json',
    jq_schema='.messages[].content',  # 加载message[]中所有的content字段
    # text_content=False  # 保持原始JSON结构, 将提取的数据转换为JSON字符串存入page_content字段中
)

# 3、加载数据
docs = json_loader.load()

# 4、显示数据
for doc in docs:
    print(doc.page_content)
```

举例4：提取04-response.json文件中嵌套在 data.items[] 里的 title、content 和 其文本

```python
# 1、导入相关依赖包
from langchain_community.document_loaders import JSONLoader
from pprint import pprint

# 2、定义JSONLoader
# # 方式一
# json_loader = JSONLoader(
#     file_path='./asset/load/04-response.json',
#     jq_schema='.data.items[].content',  # 加载04-response.json文件中嵌套在 data.items[].content 的文本
# )

# # 方式二
# json_loader = JSONLoader(
#     file_path='./asset/load/04-response.json',
#     jq_schema='.data.items[]',  # 加载04-response.json文件中嵌套在 data.items[].content 的文本
#     content_key='content',
# )

# 方式三
json_loader = JSONLoader(
    file_path='./asset/load/04-response.json',
    jq_schema='.data.items[]',  # 加载04-response.json文件中嵌套在 data.items[].content 的文本
    content_key='.title + "\\n" + .content',
    is_content_key_jq_parsable=True  # 用jq解析content_key
)

# 3、加载数据
docs = json_loader.load()

# 4、显示数据
for doc in docs:
    print(doc.page_content+'\n')
```
