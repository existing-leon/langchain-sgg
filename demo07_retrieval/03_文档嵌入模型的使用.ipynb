1、句子的向量化

```python
from langchain_openai import OpenAIEmbeddings
import os
import dotenv

# 加载环境变量
dotenv.load_dotenv()

# 给大模型相关参数赋值
os.environ['OPENAI_BASE_URL'] = os.getenv('OPENAI_BASE_URL')
os.environ['OPENAI_API_KEY'] = os.getenv('OPENAI_API_KEY')

# 创建大模型
embeddings_model = OpenAIEmbeddings(
    # model='text-embedding-ada-002'    # 1536
    model='text-embedding-3-large'  # 3072
)

# 待嵌入的文本句子
text = 'What was the name mentioned in the conversation?'

# 生成一个嵌入向量
embedded_query = embeddings_model.embed_query(text=text)

# 使用embedded_query[:5]来查看前5个元素的值
print(embedded_query[:5])

print(len(embedded_query))

print(embedded_query)
```

2、文档的向量化

文档的向量化, 接受的参数是字符串数组


举例1：

```python
from langchain_openai import OpenAIEmbeddings
import numpy as np
import pandas as pd
import os
import dotenv

# 加载环境变量
dotenv.load_dotenv()

# 配置大模型参数
os.environ["OPENAI_BASE_URL"] = os.getenv("OPENAI_BASE_URL")
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")

# 初始化嵌入模型
embeddings_modes = OpenAIEmbeddings(
    model="text-embedding-ada-002"
)

# 待嵌入的文本列表
texts = [
    "Hi there!",
    "Oh, hello!",
    "What's your name?",
    "My friends call me World",
    "Hello World!"
]

# 生成嵌入模型
embeddings = embeddings_model.embed_documents(texts)

# 打印结果
for i in range(len(texts)):
    print(f"{texts[i]}:{embeddings[i][:3]}", end="\n\n")
```
