需求：

北京明天的天气怎么样？上海呢？ （通过两次对话实现）

以ReAct模式为例来实现

```python
from langchain.agents import initialize_agent, AgentType
from langchain.memory import ConversationBufferMemory
from langchain_core.tools import Tool
# 1、导入相关的包
import os
import dotenv
from langchain_community.tools import TavilySearchResults
from langchain_openai import ChatOpenAI

# 2、加载环境变量
dotenv.load_dotenv()

# 3、创建大模型实例
os.environ['OPENAI_BASE_URL'] = os.getenv('OPENAI_BASE_URL')
os.environ['OPENAI_API_KEY'] = os.getenv('OPENAI_API_KEY')

llm = ChatOpenAI(
    model='gpt-4o-mini',
    temperature=0
)

# 3、创建搜索工具实例
os.environ['TAVILY_API_KEY'] = os.getenv('TAVILY_API_KEY')

search = TavilySearchResults(
    max_results=3
)

# 4、创建搜索工具
search_tool = Tool(
    name='Search',
    func=search.run,
    description='搜索互联网上的信息, 尤其是天气相关的'
)

# 5、获取一个记忆的实例, ConversationBufferMemory
memory = ConversationBufferMemory(
    return_messages=True,  # 返回的是消息
    memory_key='chat_history',  # 此时必须要设置memory_key, 且值必须为chat_history, 因为提示词模板中使用的是chat_history, 二者必须保持一致
)

# 5、获取AgentExecutor实例
agent_executor = initialize_agent(
    tools=[search_tool],
    llm=llm,
    agent=AgentType.CONVERSATIONAL_REACT_DESCRIPTION,  # 这个类型需要是有记忆类型的
    verbose=True,
    memory=memory
)

# 6、通过AgentExecutor调用实例, 获取响应结果
response = agent_executor.invoke(input={'input': '查询北京今天的天气情况'})

# 7、处理相应结果
print(response)
```

```python
# 6、通过AgentExecutor调用实例, 获取响应结果
response = agent_executor.invoke(input={'input': '上海呢'})

# 7、处理相应结果
print(response)
```
